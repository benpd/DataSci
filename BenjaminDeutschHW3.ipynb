```python
def run_simulation(num_particle, max_time, K, deltaT):
    temp_col = ['T_'+str(x) for x in range(max_time)]
    temp_DF = pd.DataFrame(np.sqrt(2*K*deltaT) * np.random.normal(size=(num_particle,max_time)),
                            index=range(num_particle), columns=temp_col)
    temp_DF[['T_0']]=0.00
    temp_DF = temp_DF.cumsum(axis=1)
    return temp_DF
    raise NotImplementedError

diffusion_filled_df = run_simulation(100, 1000, 2, 1)
diffusion_filled_df
```

```python
dispersion_df = diffusion_filled_df.pow(2)

dispersion_avg = dispersion_df.apply(lambda r: pd.Series({'mean': r.mean(), 'std': r.std()})).transpose()
dispersion_avg
```

```python
import scipy
from scipy.special import gamma as G

# TODO: write a function that compute g
def g(gamma, d):
    gcon = gamma**(4 / gamma)
    g1 = scipy.special.gamma((d + 2) / gamma)
    g2 = scipy.special.gamma(d / gamma)
    return (gcon * g1) / g2
```

```python
def compute_prob_distribution(absolute_position, num_cuts):
    absolute_position_copy = pd.DataFrame(absolute_position.tolist(), columns=["Particles"])
    # TODO: cut absolute_position using qcut and num_cuts
    absolute_position_copy["cut"], bins = pd.qcut( absolute_position, q = num_cuts, retbins=True )
    
    # TODO: compute deltaR
    deltaR = bins[1:] - bins[:-1]
    
    # TODO: compute the counts per cuts of the column "cut"
    counts = absolute_position_copy.groupby('cut')['Particles'].count()
    
    # TODO: normalize the counts to get the probability distribution
    pdf = counts / (deltaR * len(absolute_position_copy["Particles"].tolist()))
    
    # TODO: compute the median per cuts of the column "cut" of the absolute position
    # TODO_update: compute the median per cuts of the absolute position
    r = absolute_position_copy.groupby("cut")["Particles"].apply(np.median)
    return pdf, r

compute_prob_distribution(diffusion_filled_df3["T_999"].abs(), 50)
```

```python
def A(gamma, d):
    a = (G((d+2)/gamma)/G(d/gamma))**(gamma/2)
    return(a)

# TODO: write a function that computes B
def B(gamma, d):
    b = np.log(gamma*(G((d+2)/gamma)**(d/2))/(G(1/gamma)**((d+2)/2)))
    return(b)

# TODO: write a function that computes the dispersion
def dispersion(gamma, d, t, D):
    gdt = g(gamma,d)*(D*t)**(2/gamma)
    return(gdt)

# TODO: write a function that computes the theoretical PDF
def pdf(gamma, d, r, t, D):
    num1 = (1/dispersion(gamma,d,t,D))**(d/2)
    num2 = np.exp(-A(gamma,d)*((r/np.sqrt(dispersion(gamma,d,t,D)))**gamma)+B(gamma,d))
    return(num1*num2)  
```

```python
# TODO: compute the estimated pdf and use the r array to show the theoretical one
est_pdf, r = compute_prob_distribution(diffusion_filled_df3["T_999"].abs(),160)
est_pdf= est_pdf.tolist()
r =r.tolist()
theo_pdf = pdf(gamma=2, d=1, r=r, D=2, t=1000)

plt.figure()
plt.plot(est_pdf, label="esitmate")
plt.plot(theo_pdf, label= "theoritical")
plt.legend(loc="upper right")
```

```python
# TODO:Plot the rescaled estimated pdf for t = 199, 399, 599, 799, 999
est_pdf1, r_1 = compute_prob_distribution(diffusion_filled_df3["T_199"].abs(),160)
est_pdf2, r_2 = compute_prob_distribution(diffusion_filled_df3["T_399"].abs(),160)
est_pdf3, r_3 = compute_prob_distribution(diffusion_filled_df3["T_599"].abs(),160)
est_pdf4, r_4 = compute_prob_distribution(diffusion_filled_df3["T_799"].abs(),160)
est_pdf5, r_5 = compute_prob_distribution(diffusion_filled_df3["T_999"].abs(),160)

def x(r, gamma, t):
    dis = dispersion(gamma = gamma, d=1, t=t, D=2)
    value = (r/(np.sqrt(dis)))**gamma
    return (value)

def y(gamma, d, t, D, r, pdf):
    dis = dispersion(gamma=gamma, d=d, t=t, D=D) **(d/2)
    return (np.log(pdf*dis))

x_1 = x(r=r_1, gamma = 2, t=199)
y_1 = y(gamma = 2, d=1, t=199, D=2, r=r_1, pdf=est_pdf1)
x_2 = x(r=r_2, gamma = 2, t=399)
y_2 = y(gamma = 2, d=1, t=399, D=2, r=r_2,pdf=est_pdf2)
x_3 = x(r=r_3, gamma = 2, t=599)
y_3 = y(gamma = 2, d=1, t=599, D=2, r=r_3,pdf=est_pdf3)
x_4 = x(r=r_4, gamma = 2, t=799)
y_4 = y(gamma = 2, d=1, t=799, D=2, r=r_4, pdf=est_pdf4)
x_5 = x(r=r_5, gamma = 2, t=999)
y_5 = y(gamma = 2, d=1, t=999, D=2, r=r_5, pdf=est_pdf5)
plt.figure()
plt.plot(x_1, y_1, label = "t = 199", lw = 2)
plt.plot(x_2, y_2, label = "t = 399", lw = 2)
plt.plot(x_3, y_3, label = "t = 599", lw = 2)
plt.plot(x_4, y_4, label = "t = 799", lw = 2)
plt.plot(x_5, y_5, label = "t = 799", lw = 2)

plt.ylabel(r"$ln(P(r,t) \langle r^{2}(t) \rangle ^{\frac{1}{2}}  $", fontsize=25)
plt.xlabel(r"$(\frac{r}{\sqrt{ \langle r^2(t) \rangle}}) ^{\gamma}$", fontsize=25)
plt.legend(loc="upper right")
```

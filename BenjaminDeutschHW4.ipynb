```python
import pandas as pd
import numpy as np

N = 50
initial_spins_df = pd.DataFrame(np.ones((N,N)))
initial_spins_df
```

```python
# TODO: melt the initial_spins dataframe into a long dataframe
r_df = initial_spins_df.reset_index()
df_m = pd.melt(frame=r_df,id_vars=['index'])
# TODO: reset the current index
df_m.rename(columns={"index":"X","variable":"Y","value":"spin_value"},inplace=True)
# TODO: reset the current index
initial_spins_melted = df_m.reset_index()
# TODO: set the multiindex to be "X", "Y"
initial_spins_melted.set_index(["X","Y"],inplace=True)

initial_spins_melted
```

```python
#Definitions to locate index and assign corresponding new index
def index_up(n):
    x = n % N
    y = n // N
    if y == 0:
        y = N-1
    else:
        y = y - 1
    new_index = x + y*N
    return new_index

def index_down(n):
    x = n % N
    y = n // N
    if y == N-1:
        y = 0
    else:
        y = y + 1
    new_index = x + y*N
    return new_index

def index_left(n):
    x = n % N
    y = n // N
    if x == 0:
        x = N-1
    else:
        x = x -1
    new_index = x + y*N
    return new_index

def index_right(n):
    x = n % N
    y = n // N
    if x == N-1:
        x = 0
    else:
        x = x + 1
    new_index = x + y*N
    return new_index
```

```python
# TODO: create the columns "up_spin", "down_spin", "left_spin" and "right_spin" with the index values
# of the neighboring spins

initial_spins_melted["up_spin"] = initial_spins_melted['index'].apply(index_up)
initial_spins_melted["down_spin"] = initial_spins_melted['index'].apply(index_down)
initial_spins_melted["left_spin"] = initial_spins_melted['index'].apply(index_left)
initial_spins_melted["right_spin"] = initial_spins_melted['index'].apply(index_right)

initial_spins_melted
```

```python
# TODO: replace the missing values in "up_spin", "down_spin", "left_spin" and "right_spin" such that the periodic boundary counditions are satisfied
pos = initial_spins_melted.index.values
up_spin = [int(i) + ((N*N)-N) if pd.isnull(initial_spins_melted.loc[pos[int(i)],"up_spin"]) else initial_spins_melted.loc[pos[int(i)],"up_spin"] for i in initial_spins_melted["index"].tolist()]
down_spin = [int(i) - ((N*N)-N) if pd.isnull(initial_spins_melted.loc[pos[int(i)],"down_spin"]) else initial_spins_melted.loc[pos[int(i)],"down_spin"] for i in initial_spins_melted["index"].tolist()]
left_spin = [int(i) + (N-1) if pd.isnull(initial_spins_melted.loc[pos[int(i)],"left_spin"]) else initial_spins_melted.loc[pos[int(i)],"left_spin"] for i in initial_spins_melted["index"].tolist()]
right_spin = [int(i) - (N-1) if pd.isnull(initial_spins_melted.loc[pos[int(i)],"right_spin"]) else initial_spins_melted.loc[pos[int(i)],"right_spin"] for i in initial_spins_melted["index"].tolist()]

initial_spins_melted["up_spin"]= up_spin
initial_spins_melted["down_spin"]= down_spin
initial_spins_melted["left_spin"]= left_spin
initial_spins_melted["right_spin"]= right_spin
initial_spins_melted = initial_spins_melted.astype({"up_spin":int,"down_spin":int,"left_spin":int,"right_spin":int})
initial_spins_melted.reset_index(inplace=True)
initial_spins_melted.drop("index",axis=1,inplace=True)

initial_spins_melted
```

```python
def initialize_spin_lattice(N):
    df = pd.DataFrame(data=np.ones((N,N)))
    r_df = df.reset_index()
    df_m = pd.melt(frame=r_df,id_vars=["index"])
    df_m.rename(index=str,columns={"index":"X","variable":"Y","value":"spin_value"},
                           inplace=True)
    df_melt = df_m.reset_index()
    df_melt.set_index(["X","Y"],inplace=True)
    
    df_melt["up_spin"] = df_melt.groupby(level=0)["index"].shift(1)
    df_melt["down_spin"] = df_melt.groupby(level=0)["index"].shift(-1)
    df_melt["left_spin"] = df_melt.groupby(level=1)["index"].shift(1)
    df_melt["right_spin"] = df_melt.groupby(level=1)["index"].shift(-1)

    pos = df_melt.index.values
    up_spin = [int(i) + ((N*N)-N) if pd.isnull(df_melt.loc[pos[int(i)],"up_spin"]) else df_melt.loc[pos[int(i)],"up_spin"] for i in df_melt["index"].tolist()]
    down_spin = [int(i) - ((N*N)-N) if pd.isnull(df_melt.loc[pos[int(i)],"down_spin"]) else df_melt.loc[pos[int(i)],"down_spin"] for i in df_melt["index"].tolist()]
    left_spin = [int(i) + (N-1) if pd.isnull(df_melt.loc[pos[int(i)],"left_spin"]) else df_melt.loc[pos[int(i)],"left_spin"] for i in df_melt["index"].tolist()]
    right_spin = [int(i) - (N-1) if pd.isnull(df_melt.loc[pos[int(i)],"right_spin"]) else df_melt.loc[pos[int(i)],"right_spin"] for i in df_melt["index"].tolist()]
    
    df_melt["up_spin"]=up_spin
    df_melt["down_spin"]=down_spin
    df_melt["left_spin"]=left_spin
    df_melt["right_spin"]=right_spin
    df_melt = df_melt.astype({"up_spin":int,
                              "down_spin":int,
                              "left_spin":int,
                              "right_spin":int})
    df_melt.reset_index(inplace=True)
    df_melt.drop("index",axis=1,inplace=True)
    return(df_melt)
    
initialize_spin_lattice(50)
```

```python
def one_step(df,       # the dataframe to update
             T,        # the temperature 
             h,        # the magnetic field
             frac,     # the fraction of spins from the lattice to choose
             parityX,  # parameter to avoid to choose neighboring spins in the same batch
             parityY   # parameter to avoid to choose neighboring spins in the same batch
            ):
    
    # This part of the code is to ensure that spins chosen cannot be neighbors
    sample_spins = spin_lattice.loc[(spin_lattice["X"] % 2 == parityX) & (spin_lattice["Y"] % 2 == parityY)]
    # we sample the resulting data frame
    sample_spins = sample_spins.sample(frac=frac)  
    
    ## TODO: compute delta_e for all the spins in sample_spins. delta_e is a vector
    sum_s_j = [spin_lattice.loc[x,"spin_value"].sum()+h for x in sample_spins.loc[:,["up_spin","down_spin","left_spin","right_spin"]].values]

    n_series = pd.Series(sum_s_j,index=sample_spins.index.values)
    delta_e = (2*sample_spins["spin_value"]).multiply(n_series)

    ## TODO: Choose to flip the spins or not based on delta_e, T, and uniform random numbers
    for i,value in zip(delta_e.index.values,delta_e.values):
        if value <= 0:
            df.loc[i,"spin_value"] *= -1
    else:
        if np.exp(-value/T) > np.random.uniform():
            df.loc[i,"spin_value"] *= -1
    ## No need to return anything
```

```python
## The function to reach thermodynamic equilibrium for each value of T and h

def get_to_equilibrium(df,# the dataframe to update
                       T,# the temperature 
                       h,# the magnetic field
                       frac=0.8,# the fraction of spins from the lattice to choose
                       iteration=500 # the number of iteration of the one_step function
                      ):
    for i in range(iteration):
        
        # we choose at random 0 or 1 to select a sublattice with no neighboring spins
        parity = np.random.choice([0,1], size=2)
        
        # we apply the one_step function
        one_step(df, T, h, frac, parity[0], parity[1]) 
```

```python
## TODO: write a function to compute the average energy of the spin system
def compute_energy_mean(x, df):
    prod_spin_sum = [x.loc[i].sum() for i in df.loc[:,["up_spin","down_spin","left_spin","right_spin"]].values]
    ser = pd.Series(prod_spin_sum,index=df.index.values)
    final = (-x/2).multiply(ser)
    return(final.mean())
## TODO: write a function to compute the magnetization of the spin system
def compute_magnetization(x):
    return(x.mean())
    

## TODO: write a function to compute the specific heat of the spin system
def compute_specific_heat(x, df):
    beta = (1/x.name)
    prod_spin_sum = [x.loc[i].sum() for i in df.loc[:,["up_spin","down_spin","left_spin","right_spin"]].values]
    ser = pd.Series(prod_spin_sum,index=df.index.values)
    final = (-x/2).multiply(ser)
    return ((beta/x.name)*final.var())

## TODO: write a function to compute the magnetic susceptibility of the spin system
def compute_magnetic_susceptibility(x):
    beta = (1/x.name)
    return (beta*x.var())

statistics_df = pd.DataFrame(columns=["Energy", "Magnetization", "Specific_heat", "Magnetic_susceptibility"],
                            index=temperatures)

statistics_df["Energy"] = spin_evolution.apply(compute_energy_mean, args=(spin_lattice,))
statistics_df["Magnetization"] = spin_evolution.apply(compute_magnetization)
statistics_df["Specific_heat"] = spin_evolution.apply(compute_specific_heat, args=(spin_lattice,))
statistics_df["Magnetic_susceptibility"] = spin_evolution.apply(compute_magnetic_susceptibility)

statistics_df
```
